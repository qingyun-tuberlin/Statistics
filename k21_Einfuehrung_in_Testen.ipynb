## Begriffe

Stichprobe
- 样本
- eine Teilmenge der Grundgesamtheit ie gezogen wird, um Informationen über die gesamte Population zu erhalten

Einstichprobe
- 单样本, 指我们只使用一个样本数据集进行统计推断或假设检验。
- ein einziger Stichprobe

Zweistichproben
- Zweistichprobe, 指有 两组样本数据，我们要比较它们之间是否存在显著差异。
- 指有 两组样本数据，我们要比较它们之间是否存在显著差异。

Mehrstichprobe
- 多样本检验, 指有 三个及以上样本组（k ≥ 3），要比较它们之间是否存在显著差异。
- Anzahl der Stichprobe >= 3


Vermutung über 
- einen Parameter in der Grundgesamtheit
- eine Verteilung in der Grundgesamtheit
- das Verhalten des Untersuchungsmerkmals in der Grundgesamtheit

Hypothese
- Vermutungen über einen Parameter oder eine Verteilung in der Grundgesamtheit. Derartige
Vermutungen heißen Hypothesen

Testen 
- Ein Test ist ein Verfahren, mit dem man anhand einer Stichprobe Hypothesen überprüfen kann.

Testen von Hypothese
- Überprüfen, ob Hypothesen/Annahmen/Vermutungen zutreffen oder nicht.

Statistische Testen
- ist ein Begriff
- Die Überprüfung ∈ Statistische Testen

Statistische Tests
- Die Regeln, die zur Überprüfung eingesetzt werden
- Klassifikation nach
    - Anzahl der Stichproben
        - Einstichproben-Tests versus Zweistichproben-Tests
    - Abhängigkeit von Verteilungsannahmen
        - parametrische versus verteilungsfreie (nicht- parametrische) Tests
    - Inhalt der Hypothesen
        - Erwartungswerte, Varianzen, Anteile; Anpassungs-, Unabhängigkeitstests
    - Verteilung der Prüfstatistik
    - Hypothesenformulierung
        - zweiseitiger Test, einseitiger Test
    - Punkthypothesen und Intervallhypothesen

- Entscheidungen über die Verteilung in der Grundgesamtheit

Signifikanzniveau α
- 标准差，用于衡量一组数据与平均值的偏离程度。标准差越大，数据波动越大；标准差越小，数据越集中在平均值附近。
- wie stark die Werte einer Verteilung um ihren Mittelwert streuen. Eine große σ bedeutet starke Streuung, eine kleine σ bedeutet enge Konzentration.

Einseitige Tests
- 单侧检验：只考虑一种方向的偏离，例如“是否大于目标值”或“是否小于目标值”
- Beim einseitigen Test wird nur eine Richtung der Abweichung geprüft – entweder nach oben oder nach unten.
- E.g. H₀: μ ≤ μ₀, H₁: μ > μ₀

Zweiseitige Tests
- 双侧检验：考虑两种方向的偏离，即“是否与目标不同（无论高还是低）”
- Beim zweiseitigen Test wird geprüft, ob der Mittelwert in irgendeine Richtung von H₀ abweicht.
- E.g. H₀: μ = μ₀, H₁: μ ≠ μ₀

Nullhypothese H0
- was wir annehmen wollen

Alternativhypothese H1
- was wir verwerfen wollen, H1 ist Gegenteil von H0

Wahrscheinlichkeiten von möglichen (Fehl-) Entscheidungen
|  | H₀ wahr | H₀ falsch |
|-----------|----------|----------|
|  H₀ annehmen | ✅, 1-α  | ❌ Fehler 2. Art,  β|
| H₀ verwerfen | ❌ Fehler 1. Art,  α | ✅,  1-β |



## Einseitige und Zweiseitige Tests


| 类型 | 假设形式 | 检验方向 | 拒绝域位置 | 例子 |
|------|------------|------------|---------------|--------|
| 双侧检验 (zweiseitig) | H₀: μ = μ₀ <br> H₁: μ ≠ μ₀ | 双方向（高或低） | 两边（±z₍1−α/2₎） | 检查平均值是否偏离500 |
| 右侧检验 (einseitig rechts) | H₀: μ ≤ μ₀ <br> H₁: μ > μ₀ | 向右偏离 | 右侧（z₍1−α₎） | 检查是否超标 |
| 左侧检验 (einseitig links) | H₀: μ ≥ μ₀ <br> H₁: μ < μ₀ | 向左偏离 | 左侧（−z₍1−α₎） | 检查是否不足 |



假设啤酒厂目标装量 μ₀ = 500 ml，σ = 1.5 ml，n = 25。

| 检验类型 | 假设 | 目的 | 拒绝 H₀ 的情况 |
|------------|-------------------|----------------|----------------------|
| 双侧检验 | H₀: μ = 500 <br> H₁: μ ≠ 500 | 检查是否偏离目标 | 平均值远离 500（高或低） |
| 右侧检验 | H₀: μ ≤ 500 <br> H₁: μ > 500 | 检查是否装太多 | 平均值显著高于 500 |
| 左侧检验 | H₀: μ ≥ 500 <br> H₁: μ < 500 | 检查是否装太少 | 平均值显著低于 500 |
